## Expected value of an infinite sum
Starting from 1, in the $n^{th}$ minute a frog jumps either to the right or the left, with the following conditions:
* It jumps $\frac{1}{n}$ to the left with probability $\frac{1}{2n}$
* It jumps $\frac{2}{n}$ to the right with probability $\frac{1}{2n}$
* It stays where it was with the remaining probability i.e with $1 - \frac{1}{n}$

What is the expected position of the frog after infinite jumps?

```python
import numpy as np
def jump(n):
    threshold = np.random.uniform(low = 0, high = 1, size = 1)
    if threshold < 1 / (2*n) :
        return -1 / n
    elif threshold < 1 / n :
        return 2 / n
    else:
        return 0
```

```python
def simulation():
    z = 0
    for i in range(1,10000):
        z += jump(i)
    return z
```

```python
z = []
for i in range(1,1000):
    z.append(simulation())
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,1000), z, 'r-')
plt.title("The sum of the frog's jumps does not converge")
plt.xlabel("Different trajectories")
plt.ylabel("The frog's end position")
plt.axhline(y = np.mean(z))
plt.show()
```

```python
import math
EmpirialExpectedValue = np.mean(z)
TheoriticalExpectedValue = math.pi**2/12
EmpirialExpectedValue - TheoriticalExpectedValue
```

```python
EmpirialExpectedValue
```

```python
TheoriticalExpectedValue
```

Even after 10 000 step, the frog's position is changing significantly, which means that it does not converge almost surely (if it did, it would be almost a constant line on plot). However, it has an expected value, which is $\frac{\pi^2}{12}$. This can be shown with Fubini's theorem, by seeing that the end position of the frog is an $L_{1}$ function.
