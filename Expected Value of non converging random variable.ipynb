## Expected value of an infinite sum
Starting from 1, in the $n^{th}$ minute a frog jumps either to the right or the left, with the following conditions:
* It jumps $\frac{1}{n}$ to the left with probability $\frac{1}{2n}$
* It jumps $\frac{2}{n}$ to the left with probability $\frac{1}{2n}$
* It stays where it was with the remaining probability

What is the expected position of the frog after infinite jumps?

```python
import numpy as np
def jump(n):
    th = np.random.uniform(low = 0, high = 1, size = 1)
    if th < 1 / (2*n) :
        return -1 / n
    elif th < 1 / n :
        return 2 / n
    else:
        return 0
```

```python
def simulation():
    z = 0
    for i in range(1,10000):
        z += jump(i)
    return z
```

```python
z = []
for i in range(1,1000):
    z.append(simulation())
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,1000), z, 'r-')
plt.title("The end position of  the frog after 10 000 step")
plt.xlabel("Different trajectories")
plt.axhline(y = np.mean(z))
plt.show()
```

```python
import math
EmpirialExpectedValue = np.mean(z)
TheoriticalExpectedValue = math.pi**2/12
EmpirialExpectedValue - TheoriticalExpectedValue
```

Even after 10 000 step, the frog's position is changing significantly, which means that the Z random variable does not converge almost surely. However, it has an expected value, which is $\frac{\pi^2}{12}$.
